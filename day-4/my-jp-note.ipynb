```python
import numpy as np
from matplotlib import pyplot
x = np.arange(1, 13)
y = np.array([60, 65, 95, 75, 60, 50, 90, 80, 60, 70, 75, 50])
z = np.array([80, 75, 88, 90, 95, 78, 77, 66, 63, 68, 48, 58])
psm = np.array([45]*12)
avg = (y + z)/2
print(f"Passing Marks: {psm}")
print(f"y: {y}")
print(f"z: {z}")
print(f"Average: {avg}")
q1, = pyplot.plot(x,z, color='blue', marker='o', markerfacecolor='blue')
q2, = pyplot.plot(x,y, color='orange', marker='o', markerfacecolor='orange')
q3, = pyplot.plot(x,avg, color='black', marker='o', markerfacecolor='black')
q4, = pyplot.plot(x,psm, color='green', marker='o', markerfacecolor='green')
pyplot.xlim(0,13)
pyplot.ylim(40, 100)
pyplot.xlabel("Class")
pyplot.ylabel("Marks Obtained")
pyplot.legend([q1,q2,q3,q4],["Student1", "Student2", "Average", "Paasing Marks"])
pyplot.title("Mark Obtained by Students in Each class from 1 - 12th")
pyplot.grid()
pyplot.show()
```

```python
import numpy as np
from matplotlib import pyplot
x = np.arange(1, 13)
z = np.array([80, 75, 88, 90, 95, 78, 77, 66, 63, 68, 48, 58])
print(x)
print(z)
q1 = pyplot.scatter(x,z, color='red', marker='+')
pyplot.xlim(0,13)
pyplot.ylim(10, 100)
pyplot.xlabel("cls")
pyplot.ylabel("Marks Obtained")
pyplot.legend([q1],["Student1"])
pyplot.title("Mark Obtained by Student1 in Each cls from 1 - 12th")
pyplot.grid(color='green', linestyle='--', linewidth='.75')
pyplot.show()
```
